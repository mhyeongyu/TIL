```python
import numpy as np
import pandas as pd
import Regression as rr
import FinanceDataReader as fdr
from datetime import timedelta, date
from tqdm import tqdm

import matplotlib.pyplot as plt
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
def hola_sec_function(name, day, alpha):

    stock_code = pd.read_csv('./KOSPI_200.csv', dtype={'종목코드': str, '종목명': str})[['종목명', '종목코드']]
    data, code = rr.load_stocks_data(f'{name}', stock_code)
    stocks = rr.Stocks(data)
    day, pred = stocks.predict(code, int(day), alpha)
    
#     result = np.array(result[:-1])
    test = fdr.DataReader(f'{code}', date.today() - timedelta(days=365), date.today())[['Close']]
#     mape = mean_absolute_percentage_error(test, result)
    
    data = pd.DataFrame([day, pred], index=['Date', 'Close']).T.set_index('Date')
    
    return data, test
```

```python
data_0, test_1 = hola_sec_function('삼성전자', 20, 0.0)
data_3, test_1 = hola_sec_function('삼성전자', 20, 0.3)
data_5, test_1 = hola_sec_function('삼성전자', 20, 0.5)
data_7, test_1 = hola_sec_function('삼성전자', 20, 0.7)
data_1, test_1 = hola_sec_function('삼성전자', 20, 1)

plt.figure(figsize=(10, 6))
plt.plot(test_1.iloc[160:-19, 0], c='k', alpha=1, linewidth=2)
plt.plot(test_1.iloc[-19:, 0], c='blue', label='True', linewidth=2)

plt.plot(pd.to_datetime(data_0.index), data_0['Close'], '--',c='tomato', linewidth=4, label='Alpha 0')
plt.plot(pd.to_datetime(data_5.index), data_5['Close'], '--',c='orangered', linewidth=4, label='Alpha 0.5')
plt.plot(pd.to_datetime(data_1.index), data_1['Close'], '--',c='red', linewidth=5, label='Alpha 1')
plt.legend(loc= 'upper left')
plt.title('Samsung Electronics')
plt.xticks(rotation=30)
plt.show()
```
